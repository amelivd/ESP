```python
#Plot graph Figure 1 (a)
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace (2,1000,1000)
y=1/np.log(x)
plt.plot(x,y)
plt.xlabel('t')
plt.ylabel ('ln(t)')
plt.title ("Ideal path of the density of primes")
plt.show
```

```python
#Plot graph Figure 1 (b)
import numpy as np
import matplotlib.pyplot as plt

a = -0.1
b = 2
u = a + b * 1j

t = np.linspace(0, 100, 1000) 
f = np.exp(u * t)

plt.plot(t, f)
plt.xlabel('t')
plt.ylabel('f(t)')
plt.title('Graph of f(t) = e^(ut)')
plt.grid(True)
plt.show()

```

```python
#Plot graph Figure 1 (c)
import numpy as np
import matplotlib.pyplot as plt

a = 0.1
b = 2
u = a + b * 1j

t = np.linspace(0, 100, 1000)
f = np.exp(u * t)

plt.plot(t, f)
plt.xlabel('t')
plt.ylabel('f(t)')
plt.title('Graph of f(t) = e^(ut)')
plt.grid(True)
plt.show()
```

```python
#Plot of density of primes VS 1/lnx

chiliads = np.arange(1, 1001)

density = []
for chiliad in chiliads:
    lower_bound = (chiliad - 1) * 1000 + 1
    upper_bound = chiliad * 1000
    primes = [n for n in range(lower_bound, upper_bound + 1) if all(n % d != 0 for d in range(2, int(np.sqrt(n)) + 1))]
    density.append(len(primes) / 1000)
plt.plot(chiliads,density)
plt.xlabel('Successive Chiliads of 1000')
plt.ylabel('Density of Prime Numbers')
plt.title('Density of Prime Numbers in Successive Chiliads')
plt.show()
```
